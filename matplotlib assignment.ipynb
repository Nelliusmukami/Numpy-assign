```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
file_path = r"C:\Africa-Data-School-Curriculum\notebooks\data\jumta.csv"
df = pd.read_csv(file_path)
df
```

```python
#Create a line plot showing the monthly revenue over the two-year period. Customize the plot to make it publication-quality.
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Revenue'], marker='o', linestyle='-', linewidth=2, label='Revenue')
plt.title('Monthly Revenue (2020-2021)')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.grid()
plt.legend()
plt.show()
```

```python
# Bar plot of total number of orders per month
plt.figure(figsize=(10, 5))
plt.bar(df['Month'], df['Orders'], color='orange', label='Orders')
plt.title('Monthly Orders (2020-2021)')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend()
plt.show()
```

```python
# Scatter plot of number of orders vs average order value
plt.figure(figsize=(10, 5))
plt.scatter(df['Orders'], df['Average_Order_Value'], color='green', label='Order Value')
plt.title('Number of Orders vs Average Order Value (2020-2021)')
plt.xlabel('Number of Orders')
plt.ylabel('Average Order Value')
plt.grid()
plt.legend()
plt.show()
```

```python
# Create a subplot with all three visualizations
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(12, 18))

# Line plot of monthly revenue
ax1.plot(df['Month'], df['Revenue'], marker='o', linestyle='-', linewidth=2, label='Revenue')
ax1.set_title('Monthly Revenue (2020-2021)')
ax1.set_xlabel('Month')
ax1.set_ylabel('Revenue')
ax1.tick_params(axis='x', rotation=45)
ax1.grid()
ax1.legend()
plt.show()
```

```python
#How did the number of orders change each month during the two-year period?
# Bar plot of total number of orders per month
ax2.bar(df['Month'], df['Orders'], color='orange', label='Orders')
ax2.set_title('Monthly Orders (2020-2021)')
ax2.set_xlabel('Month')
ax2.set_ylabel('Number of Orders')
ax2.tick_params(axis='x', rotation=45)
ax2.grid(axis='y')
ax2.legend()
plt.show()
```

```python
# Is there any relationship between the number of orders and the average order
# Scatter plot of number of orders vs average order value
ax3.scatter(df['Orders'], df['Average_Order_Value'], color='green', label='Order Value')
ax3.set_title('Number of Orders vs Average Order Value (2020-2021)')
ax3.set_xlabel('Number of Orders')
ax3.set_ylabel('Average Order Value')
ax3.grid()
ax3.legend()
```

```python
# Save the subplot as a high-resolution image file.
plt.tight_layout()
plt.savefig('sales_data_analysis.png', dpi=300)
plt.show()
```
